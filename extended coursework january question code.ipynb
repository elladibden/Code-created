```python
import numpy as np
import matplotlib.pyplot as plt
import math

G = 1
c = 1
M = 1
L = 5
r = np.linspace (0.1,80.,1000)
Vn = (-((G*M)/r)+L**2/(2*(r**2)))
#print('the values of Vn are',Vn)

Vrel = (-((G*M)/r)+L**2/(2*(r**2))-(G*M*(L**2))/((c**2)*(r**3)))
#print('the values of Vrel are',Vrel)

plt.plot(r,Vn, label= 'Vn')
plt.plot(r,Vrel, label= 'Vrel')
plt.ylim(-.1,.2)
plt.legend(('Vn', 'Vrel'), fontsize='14', loc='upper right')
```

```python
labels   = ['A','B','C','D','E']
l=[3.,3.25,3.5,4.,4.5,]
for n in l:
    Vrelx = (-((G*M)/r)+(n**2)/(2*(r**2))-(G*M*(n**2))/((c**2)*(r**3)))
    plt.plot(r,Vrelx,label = 'for Vrel $L$ = {}'.format(n))
    plt.ylim(-.15,.2,)
    plt.legend (loc = 'upper right')
    
l = 2
Vnx = (-((G*M)/r)+l**2/(2*(r**2)))
plt.plot(r,Vnx,'k:', label = 'for Vn L=2')
plt.legend(loc = 'upper right')

l = 5
Vnx = (-((G*M)/r)+l**2/(2*(r**2)))
plt.plot(r,Vnx,'k--', label = 'for Vn L=5')
plt.legend(loc = 'upper right')
```

```python
p = (2*np.pi)
phi = np.linspace(0,p,1000)
r0 = 1
e = 0.5
rn = r0/(1+e*np.cos(phi))
#print(rn)
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(phi, rn)
ax.set_rlabel_position(-22.5)  # Move radial labels away from plotted line
ax.grid(True)

ax.set_title("A plot of Rn fron 0 to 2pi", va='bottom') 
plt.show()
```

```python
relp = 9.*np.pi
relphi = np.linspace(0.,relp,1000)
r0 =1 
e = 0.5
w = 0.9
rrel = r0/(1+e*np.cos(w*(relphi)))
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})
ax.plot(relphi, rrel)
ax.set_rticks([0.5, 1., 1.5, 2.,2.5])
ax.set_rlabel_position(-22.5)  # Move radial labels away from plotted line
ax.grid(True)

ax.set_title("plot of Rrel from 0 to 9 pi", va='bottom') 
plt.show()
```

```python
#make 3D graph
from mpl_toolkits import mplot3d
relp = 9.*np.pi
relphi = np.linspace(0.,relp,1000)
r0 =1 
e = 0.5
w = 0.9
rrel = r0/(1+e*np.cos(w*(relphi)))
x = rrel*np.cos(relphi)
y = rrel*np.sin(relphi)
z = relphi

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot(x,y,z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```
